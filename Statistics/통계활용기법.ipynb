```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()

# 데이터를 처리하기 위해 로드한 데이터 셋을 데이터프레임으로 변환
import pandas as pd
import numpy as np

iris_df = pd.DataFrame(data=iris_dataset.data, columns = iris_dataset.feature_names)

for f_name in iris_df.columns[:] :
  print(f_name + "의 평균값은", iris_df[f_name].mean())
  print(f_name + "의 중앙값은", iris_df[f_name].median())
  print(f_name + "의 최솟값은", iris_df[f_name].min())
  print(f_name + "의 최댓값은", iris_df[f_name].max())
  print(f_name + "의 최빈값은", iris_df[f_name].mode())
  print(f_name + "의 분산은", iris_df[f_name].var())
  print(f_name + "의 표준편차는", iris_df[f_name].std())
  print(f_name + "의 비대칭도(왜도)는", iris_df[f_name].skew())
  print(f_name + "의 첨도는", iris_df[f_name].kurt())
  print("\n")
```

```python
from scipy import stats
import pandas as pd
titanic_df = pd.read_csv("/content/drive/MyDrive/train.csv")
titanic_df.head(5)
```

```python
# 샤피로 윌크 검정(shapiro-Wilk Test)을 통한 정규성 검정
# 귀무가설이 정규성이다 vs 대립가설이 정규성이 아니다

shapiroTestResult = stats.shapiro(titanic_df["Age"])
print(shapiroTestResult)
```

```python
# 분산분석의 가정이 iid(정규성을 따라야 하므로 이러한 분석을 먼저 진행)
# 생존자와 비생존자 데이터프레임 생성
titanic_survivor = titanic_df[titanic_df["Survived"]==1]
titanic_non_survivor = titanic_df[titanic_df["Survived"]==0]

print("titanic_survivor and non-survivor 데이터 프레임\n")
print(titanic_survivor)
print(titanic_non_survivor)

for f_name in titanic_df.columns :
  if titanic_df[f_name].dtype != type(str) :
    print("Shapiro 정규성 검정 결과(statistic, p-value) -" + f_name)
    print("titanic_survivor:", stats.shapiro(titanic_survivor[f_name]))
    print("titanic_non_survivor:", stats.shapiro(titanic_non_survivor[f_name]))
    print("\n")
```

```python
# Levene 등분산 검정 실습(두 개 표본집단의 평균이 다른지 검정)

titanic_survivor = titanic_df[titanic_df["Survived"]==1]
titanic_non_survivor = titanic_df[titanic_df["Survived"]==0]

LeveneTestResult = stats.levene(titanic_survivor["Age"], titanic_non_survivor["Age"])
print(LeveneTestResult, "\n")

for f_name in titanic_df.columns :
  if titanic_survivor[f_name].dtype != type(str) :
    print("\nLevene 등분산 검정 결과 -", f_name, "\n", stats.levene(titanic_survivor[f_name], titanic_non_survivor[f_name]))
```

```python
# 등분산 / 이분산 단순 확인
print("\n[속성별 등분산/이분산 현황]\n")
for f_name in titanic_df.columns :
  if titanic_survivor[f_name].dtype != type(str) and stats.levene(titanic_survivor[f_name], titanic_non_survivor[f_name]).pvalue > 0.05 :
    print(f_name, ":등분산\n")
  elif titanic_survivor[f_name].dtype != type(str) :
    print(f_name, ":이분산\n")
```

```python
# 정규성이 있어야 t검정 가능
# 독립 t-검정

titanic_survivor = titanic_df[titanic_df["Survived"]==1]
titanic_non_survivor = titanic_df[titanic_df["Survived"]==0]

t_Test_Result = stats.ttest_ind(titanic_survivor["Age"], titanic_non_survivor["Age"], equal_var=True)
print("독립 t-검정 결과 - ", "Age")
print(t_Test_Result)


for f_name in titanic_df.columns :
  if titanic_survivor[f_name].dtype != type(str) and f_name in ["Age", "SibSp"] :
    # 2-sample t-test(등분산 가정)
    print("\n독립 t 검정 결과(등분산 - ",f_name, "\n", stats.ttest_ind(titanic_survivor[f_name], titanic_non_survivor[f_name], equal_var=True))
  elif titanic_survivor[f_name].dtype != type(str) :
    # Welch's t-test(등분산 가정 없음)
    print("\n독립 t 검정 결과(이분산 - ",f_name, "\n", stats.ttest_ind(titanic_survivor[f_name], titanic_non_survivor[f_name], equal_var=False))
```

```python
# 이원분산분석 실습
dat = pd.read_csv("/content/drive/MyDrive/poison.csv")
dat.head()

# 종속변수는 time, 독립변수는 poison, treat
dat.groupby("poison").agg(len) # poison 요인으로 구분한 집단별 표본수 = 16
dat.groupby("treat").agg(len) # treat 요인으로 구분한 집단별 표본수 = 12
dat.groupby(["poison","treat"]).agg(len) # poison과 treat 요인으로 구분한 집단별 표본수 = 4

# 이원분산분석
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

model = ols("time~C(poison) * C(treat)", dat).fit() # 교호작용 포함
tway1 = anova_lm(model)
print(tway1)

model = ols("time~C(poison) + C(treat)", dat).fit() # 교호작용 제외
tway2 = anova_lm(model)
print(tway2)
```

```python
# 단순선형회귀 실습
from sklearn.datasets import load_diabetes
datasets = load_diabetes()

diabetes_df = pd.DataFrame(data=datasets.data, columns=datasets.feature_names)
diabetes_df["target"] = datasets.target
print(diabetes_df)

from statsmodels.formula.api import ols

# 단순회귀
results = ols("target ~ bmi", data=diabetes_df).fit()
print(results.summary())

# 회귀선 시각화 및 샘플을 통해 예측된 값을 시각화
import matplotlib.pyplot as plt
sample = diabetes_df[["bmi", "target"]].sample(n=50) # 샘플생성
pred1 = results.predict(sample["bmi"]) # 샘플을 이용하여 예측을 실시
print("sample, pred1 :", sample, pred1)

plt.scatter(diabetes_df["target"], diabetes_df["bmi"], label="Origin") # 원본 데이터의 산점도 생성
plt.scatter(pred1, sample["bmi"], label="Pred1") # 예측 결과의 산점도 생성
plt.plot(results.fittedvalues, diabetes_df["bmi"], color="red", label="Regression") # 회귀직선 추가
plt.legend()
plt.show()
```

```python
from sklearn.datasets import load_iris
iris_dataset =load_iris()

from sklearn.decomposition import PCA
import pandas as pd

# 표준화 Scaler 적용
from sklearn.preprocessing import StandardScaler
iris_ss = StandardScaler().fit_transform(iris_dataset.data)
print(iris_ss[:3])

# PCA 수행
pca = PCA(n_components=2) # 주성분 수
PrincipalComponents=pca.fit_transform(iris_ss)
print(PrincipalComponents.shape)

# 주성분 누적 설명 분산 출력
print("주성분 누적 설명 분산:", pca.explained_variance_ratio_, pca.explained_variance_, sum(pca.explained_variance_ratio_))

# 데이터 프레임으로 변환
pca_df = pd.DataFrame(data=PrincipalComponents, columns=["pc1", "pc2"])
pca_df["target"] = iris_dataset.target
display(pca_df.head())

# PCA한 특성 그리기
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(pca_df, hue="target")
plt.show()
```
